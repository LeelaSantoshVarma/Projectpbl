```python
# Truth Table Generator

print("Truth Table")
print("-" * 45)
print("A\tB\tAND\tOR\tNOT A\tXOR")

values = [True, False]

for A in values:
    for B in values:
        print(A, "\t",
              B, "\t",
              A and B, "\t",
              A or B, "\t",
              not A, "\t",
              A != B)
```
