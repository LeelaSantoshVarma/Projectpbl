```python
# Truth Table Generator

print("Truth Table")
print("-" * 45)
print("A\tB\tAND\tOR\tNOT A\tXOR")

values = [True, False]

for A in values:
    for B in values:
        print(A, "\t",
              B, "\t",
              A and B, "\t",
              A or B, "\t",
              not A, "\t",
              A != B)
```

```python
# Truth Table using User Input

A = input("Enter A (True/False): ") == "True"
B = input("Enter B (True/False): ") == "True"

print()

print("A AND B =", A and B)
print("A OR B =", A or B)
print("NOT A =", not A)
print("A XOR B =", A != B)
print("A -> B =", (not A) or B)
print("A <-> B =", A == B)
```

```python
# Set Operations

A = {1,2,3,4}
B = {3,4,5,6}

print("Set A =", A)
print("Set B =", B)

print()

print("Union =", A | B)
print("Intersection =", A & B)
print("Difference (A-B) =", A - B)
print("Difference (B-A) =", B - A)
print("Symmetric Difference =", A ^ B)
```

```python
# Set Operations using User Input

A = set(map(int, input("Enter elements of Set A: ").split()))
B = set(map(int, input("Enter elements of Set B: ").split()))

print()

print("Union =", A.union(B))
print("Intersection =", A.intersection(B))
print("Difference A-B =", A.difference(B))
print("Difference B-A =", B.difference(A))
print("Symmetric Difference =", A.symmetric_difference(B))
```

```python
#Membership Test
A = {10,20,30,40}

num = int(input("Enter number: "))

if num in A:
    print(num, "is Present")
else:
    print(num, "is Not Present")
```

```python
#Subset and Superset
A = {1,2}
B = {1,2,3,4}

print("A =", A)
print("B =", B)

print("A subset of B:", A.issubset(B))
print("B superset of A:", B.issuperset(A))
```

```python
#Boolean Circuit Simulator--AND GATE
A = int(input("Enter A (0/1): "))
B = int(input("Enter B (0/1): "))

output = A and B

print("AND Gate Output =", output)
```

```python
#OR Gate
A = int(input("Enter A (0/1): "))
B = int(input("Enter B (0/1): "))

output = A or B

print("OR Gate Output =", output)
```

```python
#NOT Gate
A = int(input("Enter A (0/1): "))

output = int(not A)

print("NOT Gate Output =", output)
```

```python
#XOR Gate
A = int(input("Enter A (0/1): "))
B = int(input("Enter B (0/1): "))

output = A ^ B

print("XOR Gate Output =", output)
```

```python
#NAND Gate
A = int(input("Enter A (0/1): "))
B = int(input("Enter B (0/1): "))

output = int(not (A and B))

print("NAND Gate Output =", output)
```

```python
#NOR Gate
A = int(input("Enter A (0/1): "))
B = int(input("Enter B (0/1): "))

output = int(not (A or B))

print("NOR Gate Output =", output)
```

```python
#XNOR Gate
A = int(input("Enter A (0/1): "))
B = int(input("Enter B (0/1): "))

output = int(A == B)

print("XNOR Gate Output =", output)
```

```python
#Complete Logic Gate Simulator
A = int(input("Enter A (0/1): "))
B = int(input("Enter B (0/1): "))

print("\nLogic Gate Outputs")
print("------------------")

print("AND  :", A and B)
print("OR   :", A or B)
print("NOT A:", int(not A))
print("NOT B:", int(not B))
print("XOR  :", A ^ B)
print("NAND :", int(not (A and B)))
print("NOR  :", int(not (A or B)))
print("XNOR :", int(A == B))
```

```python
#Mini Project: Student Club Set Analysis
python = {"Rahul","Anu","Ravi","Kiran"}
ai = {"Ravi","Anu","Meena","Sita"}

print("Python Club :", python)
print("AI Club :", ai)

print()

print("Both Clubs :", python & ai)
print("Either Club :", python | ai)
print("Only Python :", python - ai)
print("Only AI :", ai - python)
```
